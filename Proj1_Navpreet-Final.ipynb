```python
from datetime import datetime
df_item['year']=pd.DatetimeIndex(df_item['release date']).year
df_item.head()
sns.distplot(df_item['year'],kde=False)
```

```python
#Visualize how popularity of genres has changed over the years
df_releases1=df_item.drop(['movie id'],axis=1).groupby(['year']).sum()
df_releases3=df_releases1.T
df_releases3
sns.heatmap(df_releases3,annot=False,fmt="",cmap='RdYlGn')
```

```python
#Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 
dfb=pd.merge(df_data,df_item,how='left',on='movie id')
dfb.head()# to get the view of dataframe
movie_count = dfb.groupby(['movie id','movie title'])['rating'].count().sort_values(ascending=False).reset_index()
movie_count.head() # to get the essence of the dataset - top movie counts 
movie_count_criteria_100 = list(movie_count[movie_count['rating']>100]['movie id'])# selection of the movie id for count>100
movie_rating_avg = dfb.groupby(['movie id','movie title'])['rating'].mean().sort_values(ascending=False).reset_index()
movie_rating_avg[movie_rating_avg['movie id'].isin(movie_count_criteria_100)].head(25)
```
